```python
class TiendaOnline:
    def __init__(self):
        self.inventario = []       
        self.clientes = {}         
        self.ventas_totales = 0.0  

    def agregar_producto(self, nombre, precio, cantidad):
        for producto in self.inventario:
            #Si el producto ya está en el inventario sólo actualizamos la cantidad:
            if producto["nombre"] == nombre:
                producto["cantidad"] += cantidad
                break
            #En caso contrario, el producto se añade, con su inventario correspondiente:
            else:  
                nuevo_producto = {"nombre": nombre, "precio": precio, "cantidad": cantidad}
                self.inventario.append(nuevo_producto)

    def agregar_cliente(self, nombre, email):
        #Si el nombre del cliente ya está registrado nos devuelve el mensaje de que ya está registrado y por tanto no se ejecuta el bloque del registro:
        if nombre in self.clientes:
            print("Ya existe un cliente registrado con este nombre")
        #Si no, se añade y se informa al usuario de que se ha añadido correctamente:
        else:
            cliente = {"nombre": nombre, "email": email, "compras_cliente": []}
            self.clientes[nombre] = cliente
            print("El cliente ha sido registrado correctamente")

    def realizar_compra(self):
        #Creación de una lista vacía a la que se irán añadiendo los productos:
        carro_compra = []
        #Creamos un bucle con while para que el usuario pueda añadir productos hasta que decida parar ingresando la palabra "fin":
        while True:
            #En primer lugar se muestra el inventario para que el usuario vea qué productos hay disponibles:
            print("El inventario actual es el siguiente:")
            for producto in self.inventario:
                print(f'Nombre:{producto["nombre"]}; Precio: {producto["precio"]}; Cantidad: {producto["cantidad"]}')
            #Con un input pedimos al usuario que vaya añadiendo productos y cantidades o que indique si desea finalizar la compra ingresando "fin"
            seleccion_producto = input("Por favor, seleccione el producto que desea añadir al carro de la compra. Si no desea comprar más indique fin")
            if seleccion_producto == "fin":
                break
            encontrado = False
            #A continuación se verifica la existencia del producto y la cantidad indicadas por el usuario:
            for prod in self.inventario:
                if prod["nombre"] == seleccion_producto:
                    encontrado = True
                    seleccion_cantidad = float(input("Por favor, seleccione la cantidad"))
                    if seleccion_cantidad <= prod["cantidad"]:
                        carro_compra.append({"nombre": seleccion_producto, "precio": prod["precio"], "cantidad": seleccion_cantidad})
                        prod["cantidad"] -= seleccion_cantidad
                        print(f"{seleccion_cantidad} unidades de {seleccion_producto} se han añadido a su carro de la compra correctamente")
                    else:
                        print("Lo sentimos, no quedan suficientes unidades de ese producto en stock")
            #Si el producto no se encuentra en el inventario se imprime el mensaje informativo correspondiente:
            if not encontrado:
                    print(f"Disculpa, {seleccion_producto} no se econtró en el inventario")
        #Se calcula el total de la factura:
        factura = 0.0
        for producto in carro_compra:
            factura += producto["precio"] * producto["cantidad"]
            factura = round(factura, 2) #Se redondea el total de la factura a dos decimales
            print(f"el total de su compra son {factura} euros")
            return carro_compra # Se devuelve la lista de productos comprados
        



```
